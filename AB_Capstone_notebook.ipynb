```python
#KNN

#First step is grouping prices into economic(cheap), low-mid, high-mid, and high to improve the model.
#KNN regression tries to predict the value of the output variable by using a local average.
#KNN classification attempts to predict the class to which the output variable belong by computing the local probability.
```

```python
# create a list of conditions
conditions = [
    (data3['price'] < 100),
    (data3['price'] >= 100) & (data3['price'] < 250),
    (data3['price'] >= 250) & (data3['price'] < 600),
    (data3['price'] >= 600) ]

# create a list of the values that we can assign for each condition
values = ['economic', 'low-mid', 'high-mid','high']

# create a new column and use np.select to assign values to it using our lists as arguments
data3['price_range'] = np.select(conditions, values)
```

```python
data3['price_range'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
```

```python
labelencoder = LabelEncoder()# Assigning numerical values and storing in another column
data3['price_rng_NEW'] = labelencoder.fit_transform(data3['price_range'])
```

```python
data4=data3[['id','minimum_nights','availability_365', 
         'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count',
         'room_type_NEW','city_NEW', 'state_NEW', 'price', 'price_rng_NEW']]
```

```python
data4.corr().style.background_gradient(cmap='magma')
```

```python
#K-nearest neighbors (KNN)

#KNN classification 
neigh = KNeighborsClassifier(n_neighbors=3)

X1=data4.iloc[:, :-1].values
Y1=data4['price_rng_NEW'].values

# Split into training and test  
X_train, X_test, y_train, y_test = train_test_split( 
             X1, Y1, test_size = 0.4, random_state=1) 

#standardize data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

#fit the model
neigh.fit(X_train, y_train)
```

```python
# Predicted class
y_pred3=neigh.predict(X_test)
```

```python
KNNmod = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred3.flatten()})
KNNmod
```

```python
#comparison 
preds3first20=KNNmod.head(20)
c2='blue', 'dimgray'
preds3first20.plot(kind='barh',figsize=(9,6), color=c2)
plt.grid(which='major', linestyle='-', linewidth='0.3', color='orange')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#The above model did not work with the original price parameters. 
#The model does improve significantly after grouping prices. We can say the model works pretty good.
```

```python
#KNN Regression 
knn = KNeighborsRegressor(algorithm='auto')
cols = data4[['calculated_host_listings_count', 'room_type_NEW']].values

X2=cols
Y2=data4['price_rng_NEW'].values
```

```python
#normalize 
X_normalized = preprocessing.normalize(X2, norm='max')
X_normalized
```

```python
# Split into training and test  
X_train, X_test, y_train, y_test = train_test_split( 
             X_normalized, Y2, test_size = 0.3) 

#fit the model
knn.fit(X_train, y_train)
```

```python
# Predicted class
y_pred4=knn.predict(X_test)
```

```python
KNNreg = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred4.flatten()})
KNNreg
```

```python
KNNreg['squared_error'] = (KNNreg['Predicted'] - KNNreg['Actual'])**(2)
mse = KNNreg['squared_error'].mean()
rmse2 = mse ** (1/2)
rmse2
```
